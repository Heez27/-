# 3.1 넘파이 배열

```python
import numpy as np
```

```python
# 배열 생성
A = np.array([[1,2], [3,4]])
A
```

```python
# 배열의 자료형
type(A)
```

```python
# 배열의 차원
A.ndim
```

```python
# 배열의 크기
A.shape
```

```python
# 원소 자료형
A.dtype
```

```python
# 원소별 최댓값, 평균값, 최솟값, 합계 구하는 함수
A.max(), A.mean(), A.min(), A.sum()
```

```python
# 배열의 접근
A[0] # A행렬의 1번째 행
```

```python
A[1] # A행렬의 2번째 행
```

```python
# 배열 원소에 접근
print(A[0][0], A[0][1]) # 1행 1열, 1행 2열 원소 출력
print(A[1][0], A[1][1]) # 2행 1열, 2행 2열 원소 출력
```

```python
# 조건에 맞는 원소들만 인덱싱
A[A>1] # 1보다 큰 원소 출력
```

```python
# 배열 형태 바꾸기(주대각선을 기준으로 뒤바꾸는 것)
A.T # A.transpose()로 해도됨
```

```python
A.transpose()
```

```python
# 배열 평탄화
A.flatten()
```

```python
# 배열의 원소별 연산
A+A
```

```python
A-A
```

```python
A*A
```

```python
A/A
```

```python
# 브로드캐스팅: 넘파이에서는 행렬 크기가 달라도 연산 가능함 (크기가 작은 행렬을 확장시켜주므로)
A = np.array([[1,2], [3,4]])
B = np.array([10,100])

A*B
```

```python
# 내적 구하기
B.dot(B)
```

```python
A.dot(B)
```
