# Pandas
* 금융 데이터 분석을 목적으로 개발
* Series와 DataFrame 자료형 객체 제공

```python
import pandas as pd
```

```python
# 인덱스 설정x 경우 → 0부터 시작하는 정수형 인덱스가 자동으로 생성됨
s = pd.Series([0.0, 0.1, 0.2, 0.3, 0.4])
s
```

```python
# index 변경
s.index = pd.Index([5,4,3,2,1])

# index명 설정
s.index.name = 'opposite_Index' 
s
```

```python
# 시리즈명 변경
s.name = 'My_Series'
s
```

```python
# 데이터 추가
s[0] = 5.5
s
```

```python
# 인덱스 지정해서, 새로운 시리즈 생성
s2 = pd.Series([1.1, 1.2], index = [9, 10])
s2
```

```python
# s 시리즈에 ser 시리즈 추가
s = s.append(s2)
s
```

```python
# 인덱스 찾기
s.index[-1]
```

```python
# values로 찾기
s.values[-1]
```

```python
# loc 인덱서: index값에 해당하는 데이터를 표시
s.loc[10]
```

```python
# iloc 인덱서: 지정한 정수 인덱스에 해당하는 값을 표시
s.iloc[-1]
```

```python
s.iloc[:] # 전체를 출력할 때, Series로 반환
```

```python
s.values[:] # 배열로 반환
```

```python
# 데이터 삭제
s.drop(10) # s.drop(s.index[-1])과 같음
```

```python
# 시리즈 정보 보기 (describe() 메소드)
s.describe() # 원소 갯수, 평균, 표준편차, 최솟값, 제1 & 2 & 3 사분위수, 최댓값
```

```python
# 시리즈 출력하기
s = pd.Series([0.0, 3.6, 2.0, 5.8, 4.2, 8.0, 5.5, 6.7, 4.2], index = [0.0, 1.2, 1.8, 3.0, 3.6, 4.8, 5.9, 6.8, 8.0])
s.index.name = 'My_IDX' # 인덱스명 설정
s.name = 'MY_SERIES' # 시리즈명 설정

import matplotlib.pyplot as plt
plt.title("ELLIOTT_WAVE")
plt.plot(s, 'bs--') # 시리즈를 bs-- (푸른 사각형과 점선) 형태로 출력
plt.xticks(s.index) # x축의 눈금값을 인덱스값으로 설정
plt.yticks(s.values) # y축의 눈금값을 데이터값으로 설정
plt.grid(True)
plt.show()
```
